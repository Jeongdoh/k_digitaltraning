```python
import pandas as pd

df = pd.read_csv('./data/향수 수입추이.csv')
```

```python
df['시점']=['20년1분기','20년2분기','20년3분기','20년4분기','21년1분기','21년2분기','21년3분기','21년4분기','22년1분기','22년2분기','22년3분기','22년4분기']
```

```python
df
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 경로 설정
font_path = 'C:/Windows/Fonts/malgun.ttf'  # 원하는 한글 폰트 파일 경로로 수정해주세요.

# 한글 폰트 이름 얻어오기
font_name = fm.FontProperties(fname=font_path).get_name()

# Matplotlib에서 한글 폰트로 지정
plt.rc('font', family=font_name)


```

```python
df_grouped = df.groupby('시점').sum()
```

```python
plt.plot(df_grouped.index, df_grouped['향수 수입액 (천달러)'])
plt.title('향수 수입현황')
plt.xlabel('연도')
plt.ylabel('수입액')

plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
plt.bar(df_grouped.index, df_grouped['향수 수입액 (천달러)'])
plt.title('향수 수입현황')
plt.xlabel('연도')
plt.ylabel('수입액')

plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
plt.scatter(df['시점'], df['향수 수입액 (천달러)'])
plt.title('향수 수입현황')
plt.xlabel('연도')
plt.ylabel('수입액')

plt.xticks(rotation=45, ha='right')

plt.show()
```
