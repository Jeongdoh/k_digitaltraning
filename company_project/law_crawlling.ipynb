# 기본법

```python
import time
import random
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


times=random.randint(3,9)

# 크롬 드라이버 경로 설정
driver_path = 'C:\\Program Files\\Google\\Chrome\\Application\\chromedriver.exe'

# 네이버 VIBE 사이트 접속
driver = webdriver.Chrome(driver_path)
driver.get("https://www.law.go.kr/LSW/lsSc.do?section=&menuId=1&subMenuId=15&tabMenuId=81&eventGubun=060101&query=%EC%86%8C%EB%B0%A9%EA%B8%B0%EB%B3%B8%EB%B2%95#undefined")
time.sleep(2)
driver.implicitly_wait(10)

law1 = []
# 가사 추출
lyrics = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, 'bl')))
time.sleep(times)
for i in lyrics:
    law1.append(i.text)
# law.append(lyrics.text)

# 크롬 브라우저 종료
driver.quit()

# 리스트를 파일에 저장
with open('소방기본법.txt', 'w', encoding='utf-8') as f:
    for law in law1:
        f.write(law)
        f.write('\n')
```
