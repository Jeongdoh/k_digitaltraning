```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.tree import plot_tree, export_graphviz, export_text

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data Collection
iris_bunch = load_iris()
```

```python
# for _key in iris_bunch.keys():
#     print(f'[{_key}]\n{iris_bunch[_key]}')
```

```python
data = iris_bunch['data']
target = iris_bunch['target']
```

```python
data.shape, target.shape
```

```python
train_X, test_X, train_y, test_y = train_test_split(
    data, target, train_size = 0.8)
```

```python
train_X.shape, train_y.shape
```

```python
test_X.shape, test_y.shape
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(train_X, train_y)
```

```python
plot_tree(dtc)
plt.savefig('tree.jpg')
```

```python
print(export_text(dtc))
```

```python
import graphviz
```

```python
export_graphviz(dtc, out_file = 'dtc.dot', filled = True)
```
