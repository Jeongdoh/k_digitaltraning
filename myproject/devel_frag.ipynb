마지막으로 compare_models() 함수에서 출력이 없는 경우, 모든 모델의 성능을 비교한 결과에 대해 아무런 정보가 없다는 것을 의미합니다.

이 경우, 사용된 데이터의 특성, 모델 하이퍼파라미터, 그리고 compare_models() 함수에 전달된 매개변수 등을 검토하여 원인을 파악해야 합니다.


compare_models() 함수는 매개변수 없이 호출되면 모든 분류 모델에 대해 기본적인 하이퍼파라미터를 사용하여 모델을 학습하고 성능을 비교합니다.

하지만 성능을 개선하거나 비교 대상 모델을 제한하려는 경우에는 매개변수를 조정할 필요가 있습니다.

다음은 compare_models() 함수에서 자주 사용되는 일부 매개변수입니다.

exclude: 성능 비교 대상에서 제외할 모델의 이름을 지정합니다.
문자열 또는 문자열의 리스트로 지정할 수 있습니다.

include: 성능 비교 대상에 포함할 모델의 이름을 지정합니다.
문자열 또는 문자열의 리스트로 지정할 수 있습니다.

fold: k-fold 교차 검증을 사용하여 모델 성능을 계산할 때, k값을 지정합니다.
기본값은 10입니다.

round: 성능 지표를 반올림할 소수점 자리수를 지정합니다. 기본값은 4입니다.

sort: 성능 지표를 기준으로 모델의 성능을 정렬합니다. 문자열 형태로 성능 지표와 정렬 방법을 지정할 수 있습니다.
이 외에도 compare_models() 함수에서 사용할 수 있는 다양한 매개변수가 있으며,
자세한 내용은 PyCaret의 공식 문서를 참조하시기 바랍니다.

```python
import random

# 이름이 담긴 리스트
names = ['김수진' , '이아름' , '신지은' ,'손지현', '하주리' ,'이소정', '한수아' ,'이현경선생님' ,'노혜미선생님' , '석은수', '김현아' ,'김소현강사님' ,'박수빈선생님']

# 첫 번째 이름 랜덤 출력
first_name = random.choice(names)
print("First name:", first_name)

# 첫 번째 이름을 제외한 리스트
remaining_names = [name for name in names if name != first_name]


```
