```python
data=pd.read_csv('./data/ab_data.csv', usecols = [0, 1])
target_df=df['Species']
```

```python
data
```

```python
import numpy as np
data = np.array(data)
target_df = np.array(target_df)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = train_test_split(
    data, target_df, stratify = target_df, test_size = 0.2)
```

```python
kn.fit(train_input, train_target)
```

```python
kn.score(test_input, test_target)
```

```python
for k in range(1, 56):
    kn.n_neighbors = k
    kn.fit(data, target_df)
    print(f'[{k}]: {kn.score(data, target_df)}')
```
