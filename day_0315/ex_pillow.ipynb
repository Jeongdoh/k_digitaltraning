```python
from PIL import Image
```

```python
FILE='../img/flower.png'
```

```python
try:
    im_data = Image.open(FILE)
except Exception as e:
    print(e)
```

```python
type(im_data)
```

```python
# Png 이미지 파일 객체 속성 ---------------------------
im_data.width, im_data.height, im_data.size, im_data.format, im_data.mode
```

```python
# 이미지 보기 ==>  이미지 관련 프로그램
im_data.show()
```

```python
# 실제 이미지 raw 데이터 16진수로 표시
im_data.tobytes()
```

```python
im_data.resize((100, 100))
```

```python
im_data.rotate(45)
```
